```julia
# packages to download
import DataFrames
```

```julia
using HTTP

# Define the URL of the CSV file
csv_loc = "https://opendata-nzta.opendata.arcgis.com/datasets/NZTA::road-events.csv?where=1=1&outSR=%7B%22latestWkid%22%3A2193%2C%22wkid%22%3A2193%7D"

# Define the relative path for the downloaded CSV file (same directory as the script)
tempname = "RoadEvents.csv"

# Download the CSV file
response = HTTP.get(csv_loc)
```

```julia
# Check if the download was successful (HTTP status code 200 indicates success)
if response.status == 200
    # Save the downloaded CSV data to the specified file
    open(tempname, "w") do file
        write(file, String(response.body))
    end
    println("CSV file downloaded and saved as $tempname")
else
    println("Failed to download the CSV file. HTTP status code: $(response.status)")
end
```

```julia
road_events = DataFrame(CSV.File(tempname))
```

```julia
filtere_road_events = DataFrame(
    X = road_events.X,
    Y = road_events.Y,
    ObjectID = road_events.OBJECTID)
```
